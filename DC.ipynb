# Importing data

```python
import pandas as pd
pd.set_option('display.max_columns', None)
data = pd.read_parquet("yellow_tripdata_2025-01.parquet")
df = pd.DataFrame(data)
df
```

# Cleaning

## checking for vendor id and removing the minor classes

```python
df['VendorID'].value_counts()
```

```python
df = df[df["VendorID"] < 3]
```

## Filling Null Values

```python
df.isnull().mean() * 100
```

### Filling Passenger Count

```python
df["passenger_count"].value_counts()
```

```python
# Get the probabilities
prob = df['passenger_count'].value_counts(normalize=True)
prob
```

```python
import numpy as np
n_missing = df['passenger_count'].isnull().sum()

# Randomly choose values based on probability
fill_values = np.random.choice(prob.index, size=n_missing, p=prob.values)

# Fill the missing values
df.loc[df['passenger_count'].isnull(), 'passenger_count'] = fill_values
```

### Filling RateCardID

```python
df["RatecodeID"].value_counts()
```

```python
import numpy as np
df['RatecodeID'] = df['RatecodeID'].replace(99, np.nan)
```

```python
# Get the probabilities
prob = df['RatecodeID'].value_counts(normalize=True)
missing_value = df['RatecodeID'].isnull().sum()
```

```python
import numpy as np
random_fill = np.random.choice(prob.index,size=missing_value,p=prob.values)
df.loc[df["RatecodeID"].isnull(),"RatecodeID"] = random_fill
```

### Removing the store and fed flag (since it is not needed for our analysis)

```python
df = df.drop(columns=["store_and_fwd_flag"])
```

### Filling Airport Fee

```python
df["Airport_fee"].describe()
```

```python
df['Airport_fee'].value_counts(dropna=False)
```

```python
# converting all negative values to 0
df['Airport_fee'] = df['Airport_fee'].abs()
```

```python
# correcting the outliers more than 2 and also,
# there is 0.25 and 1.25 as around 10 rows only, which clearly tells it is a noise.
# so making all to 1.75
df.loc[df["Airport_fee"] > 0, "Airport_fee"] = 1.75
```

```python
# Filling as 0 since most trips are not airport.
df["Airport_fee"] = df["Airport_fee"].fillna(0)
```

```python
df['Airport_fee'].value_counts(dropna=False)
```

### Congestion Surcharge

```python
df.isnull().mean() * 100
```

```python
df["congestion_surcharge"].value_counts()
```

```python
# Less than 0 is a outlier
df.loc[df["congestion_surcharge"]<0,"congestion_surcharge"]=0
```

```python
# Get the probabilities
empty_values = df["congestion_surcharge"].isnull().sum()
Prob = df['congestion_surcharge'].value_counts(normalize=True)

import numpy as np
fill_values = np.random.choice(Prob.index,size=empty_values,p=Prob.values)
df.loc[df["congestion_surcharge"].isnull(),"congestion_surcharge"] = fill_values
```

# Feature Engineering

### Extracting Time Data

```python
df["pickup_hour"] = df["tpep_pickup_datetime"].dt.hour
df["pickup_day"] = df["tpep_pickup_datetime"].dt.day
df["pickup_weekday"] = df["tpep_pickup_datetime"].dt.weekday
df["is_weekend"] = df["pickup_weekday"].isin([5,6]).astype(int)
```

```python
df["trip_speed"] = df["trip_distance"] / (df["trip_duration"] / 60)
```

### Amount Data

```python
df["tip_percentage"] = (df["tip_amount"] / df["fare_amount"]) * 100
```

```python
df["revenue_per_mile"] = df["total_amount"] / df["trip_distance"]
```

### Speed

```python
df["trip_speed"].describe()
```

```python
# Domain Knowledge
df = df[df["trip_speed"] < 80]
df = df[df["trip_speed"] > 1]
```

### Time-Based Analysis

#### Revenue By Hour

```python
trips_by_time = df.groupby("pickup_hour")["total_amount"].sum()
```

```python
import matplotlib.pyplot as plt
plt.plot(trips_by_time.index,trips_by_time.values)
plt.xlabel("Hour")
plt.ylabel("Revenue By Million")
plt.title("Hour Based Revenue Trend - January")
plt.xticks(trips_by_time.index, rotation=90)
plt.show()
```
